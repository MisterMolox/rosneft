```
from tqdm import tqdm
from google.colab import drive
import numpy as np
from PIL import Image
from keras import backend as K
from skimage.io import imsave    

K.set_image_data_format('channels_last')
drive.mount('/content/drive')
```

```
TRAIN = 512
TRAIN_ADD = 384
PATH = '/content/drive/My Drive/Rosneft Seismic/train/'
layer = '6'

print('LOAD TRAIN DATA')
  
for k in tqdm(range(TRAIN_ADD)): 
  img = np.array(Image.open(PATH+'answer/'+layer+'/xline_'+str(k+2017)+'.png'))
  
  mask_border = np.zeros((384, 512)) 
  
  for j in range(512):
    i_up = 383
    i_down = 0
    for i in range(384):
      
      if img[i, j] == 255:
        if i_up > i: 
          i_up = i
          
      if img[i, j] == 255:
        if i_down < i: 
          i_down = i
          
    if (i_down - i_up) < 4: 
      print('WARNING', i_down - i_up, 'xline_'+str(k+2017)+'.png')
      
    mask_border[i_up, j] = 255
    mask_border[i_up + 1, j] = 255
    mask_border[i_down, j] = 255
    mask_border[i_down - 1, j] = 255
      
  mask_border = mask_border.astype(np.uint8)
  imsave(PATH+'answer/'+layer+'_border/xline_'+str(k+2017)+'.png', mask_border)
  
for k in tqdm(range(TRAIN)): 
  img = np.array(Image.open(PATH+'answer/'+layer+'/inline_'+str(k+3160)+'.png'))
  
  mask_border = np.zeros((384, 384)) 
  
  for j in range(384):
    i_up = 383
    i_down = 0
    for i in range(384):
      
      if img[i, j] == 255:
        if i_up > i: 
          i_up = i
          
      if img[i, j] == 255:
        if i_down < i: 
          i_down = i
          
    if (i_down - i_up) < 4: 
      print('WARNING', i_down - i_up, 'inline_'+str(k+3160)+'.png') 
      
    mask_border[i_up, j] = 255
    mask_border[i_up + 1, j] = 255
    mask_border[i_down, j] = 255
    mask_border[i_down - 1, j] = 255
      
  mask_border = mask_border.astype(np.uint8)
  imsave(PATH+'answer/'+layer+'_border/inline_'+str(k+3160)+'.png', mask_border)

print('END LOAD TRAIN DATA')
```
