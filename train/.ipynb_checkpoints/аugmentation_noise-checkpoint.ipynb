```python
import numpy as np
from PIL import Image
from tqdm import tqdm
from skimage.io import imsave
import matplotlib.pyplot as plt
import os 
import random
```

```python
mass_number = []
PATH = 'E:\\Google Диск/Rosneft Seismic/test/images/'
dir_list = os.listdir(PATH)
for s in dir_list:
    if s[0] == 'x': 
        num = s[6:-4]
        mass_number.append(num)
print(mass_number)
```

```python
path_1 = '/answer/'
path_2 = '/images/'
path_4 = 'E:\\Google Диск\\Rosneft Seismic\\test\\images\\'

for i in tqdm(range(384)):
    number = str(2017+i)
    img_name = 'xline_' + number + '.png'
    new_img_name = 'noisexline_' + number + '.png'
    img = np.array(Image.open(os.getcwd()+path_2 + img_name))
    
    nom = mass_number[random.randint(0, 1039)]
    
    noise_img_name = 'xline_' + nom + '.png'
    noise_img = np.array(Image.open(path_4 + noise_img_name))
    
    m = random.randint(0, noise_img.shape[1] - 512)
    new = img[0: 300, :]
       
    new = np.concatenate((new, noise_img[300 : 384, m : m+512]), axis = 0)
    imsave(os.getcwd()+path_2 + new_img_name, new)
```

```python
plt.figure(figsize=(16, 8))
plt.subplot(121)
plt.imshow(img)
plt.subplot(122)
plt.imshow(new)
plt.show()
plt.close()
```
