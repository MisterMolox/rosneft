```python
import numpy as np                                     # Numpy 1.15.4
import pandas as pd                                    # Pandas 0.23.4
import csv                                             # CSV 1.0
import os                                              # 
from utils import mask2rle, rle2mask                   # Decoder csv to picture
from skimage.io import imsave                          # 
```

```python
data = pd.read_csv("train/train_masks.csv")            # Загружаем закодированные ответы на train выборку
path = os.getcwd()                                     # Рабочая директория

for x in range(6272):                                  # Раскодируем все классы в отдельные изображения для каждой картинки
    n_class = data['ClassId'][x]
    if data['ImageId'][x].split('_')[0] == 'inline':
        image = rle2mask(data['EncodedPixels'][x], (384, 384))
    else:
        image = rle2mask(data['EncodedPixels'][x], (384, 512))
        
    image = (image[:, :] * 255.).astype(np.uint8)
    name = name = path + '\\train\\answer\\' + str(n_class) + '\\' + data['ImageId'][x]
    imsave(name, image)
    
    if (x%7 == 0): 
        print('foto number ', x//7, ' download')
```
